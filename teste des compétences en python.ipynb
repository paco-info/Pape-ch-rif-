```python
# EXERCICE 1

start_num = int(2000)
  
end_num = int(3200)

cnt = start_num
  
while cnt <= end_num:
    
    if cnt % 7 == 0 and cnt % 5!=0:
        print(cnt, " est divisible par 7 et non multiple de 5.")

    cnt += 1
```

```python
# EXERCICE 2

nbr = int(input('Entrez un nombre : '))
fact = 1
for i in range(1, nbr+1):
    fact = fact * i

print (nbr,'! = ',fact)
```

```python
# EXERCICE 6

import numpy as np
  
array1 = np.array([0, 1, 1])
array2 = np.array([2, 2, 1])
  
# Original array1
print(array1)
  
# Original array2
print(array2)
  
# Covariance matrix
print("\nCovariance matrix of the said arrays:\n",np.cov(array1, array2))
```
